### Data Loading


```python
Customers = pd.read_csv("telecom_customer_churn.csv", index_col=0)
Customers
```

```python
zip_code=pd.read_csv("telecom_zipcode_population.csv", index_col=0)
zip_code
```

```python
# Merging Zipcode with customers:
df = pd.merge(left= Customers, right= zip_code, how= 'inner', on= 'Zip Code')
df
```

```python
df['Churn_Flag'] = df['Customer Status'].apply(lambda x: 1 if x == "Churned" else 0)
```

### Impute Missing Values:

```python

from sklearn.impute import SimpleImputer
simp_imp= SimpleImputer(strategy='constant',fill_value='No')
x_train[['Multiple Lines','Online Security','Online Backup','Device Protection Plan','Premium Tech Support','Streaming TV','Streaming Movies','Streaming Music','Unlimited Data']] = simp_imp.fit_transform(x_train[['Multiple Lines','Online Security','Online Backup','Device Protection Plan','Premium Tech Support','Streaming TV','Streaming Movies','Streaming Music','Unlimited Data']])
x_test[['Multiple Lines','Online Security','Online Backup','Device Protection Plan','Premium Tech Support','Streaming TV','Streaming Movies','Streaming Music','Unlimited Data']] = simp_imp.transform(x_test[['Multiple Lines','Online Security','Online Backup','Device Protection Plan','Premium Tech Support','Streaming TV','Streaming Movies','Streaming Music','Unlimited Data']])
```

```python
from sklearn.impute import SimpleImputer
simp_imp= SimpleImputer(strategy='constant',fill_value='No internet')
x_train[['Internet Type']] = simp_imp.fit_transform(x_train[['Internet Type']])
x_test[['Internet Type']] = simp_imp.transform(x_test[['Internet Type']])
```

```python
from sklearn.impute import SimpleImputer
simp_imp= SimpleImputer(strategy='constant',fill_value=0)
x_train[['Avg Monthly GB Download','Avg Monthly Long Distance Charges']] = simp_imp.fit_transform(x_train[['Avg Monthly GB Download','Avg Monthly Long Distance Charges']])
x_test[['Avg Monthly GB Download','Avg Monthly Long Distance Charges']] = simp_imp.transform(x_test[['Avg Monthly GB Download','Avg Monthly Long Distance Charges']])
```

### Undersampling:

```python
from imblearn.under_sampling import RandomUnderSampler

under_sampler = RandomUnderSampler()

x_train_under_sample, y_train_under_sample = under_sampler.fit_resample(x_train, y_train)
y_train_under_sample.value_counts()
```

### Oversampling:

```python
from imblearn.over_sampling import RandomOverSampler

over_sampler = RandomOverSampler()

x_train_over_sample, y_train_over_sample = over_sampler.fit_resample(x_train, y_train)
y_train_over_sample.value_counts()
```

### SMOTE:

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

x_train_smote, y_train_smote = smote.fit_resample(x_train, y_train)
y_train_smote.value_counts()
```
